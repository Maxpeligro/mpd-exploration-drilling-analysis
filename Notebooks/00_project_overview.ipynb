### 1. Project Context

This project analyses multi-table geological and geochemical data from 
the MPD copper-gold porphyry project in south-central British Columbia, 
Canada, located near Merritt within the Nicola Belt. This belt hosts 
several significant porphyry copper-gold deposits. The MPD dataset was 
sourced from the publicly available BC ARIS assessment report, which 
provides drillhole logs, assay results, and geological descriptions from 
the 2023 drill program.

The goal of this project is to build a clean, interval-aligned dataset 
that integrates assay intervals with lithology, alteration, and 
mineralization logs — enabling meaningful grade distribution analysis, 
geological interpretation, spatial visualization, and predictive 
modelling. The workflow reflects real exploration challenges: inconsistent 
interval boundaries, overlapping logs, variable sampling lengths, and the 
need to merge multiple domain-specific datasets into a unified structure.

### 2. Analytical Methods (ICP-MS and Fire Assay)

Assay data for the MPD project was analysed using the following laboratory 
methods as reported in the ARIS assessment report:

- **ICP-MS (Inductively Coupled Plasma Mass Spectrometry)** — ultra-sensitive, 
  capable of detecting metals at trace concentrations (ppm or lower). Because 
  of this sensitivity, elements like copper rarely report true zero values — 
  even background host rocks typically contain measurable trace Cu.

- **Fire Assay (Au)** — industry standard for gold, providing accurate 
  ppm-level measurements.

**Implication:** The dataset contains no zero-grade copper intervals. The 
minimum Cu value (~0.00013%) reflects the instrument detection limit, not 
barren geology. This is consistent with porphyry systems, where broad halos 
of weak mineralization produce trace Cu throughout the host rocks.

### 3. Data Provenance and Sources

All data was obtained from the publicly available BC ARIS filing for the 
MPD project. The ARIS package includes multi-table drillhole datasets 
covering lithology, alteration, mineralization, and assay intervals.

The ARIS package did not include a collar file. Collar coordinates 
(Easting, Northing, Elevation, Azimuth, Dip, and Total Depth) were 
extracted from the drillhole collar table in the main ARIS PDF report 
and manually transcribed into a structured table for this project.

The primary interval-based tables used in this workflow are:

- **Assay** — multi-element ICP-MS geochemistry and gold fire assay results
- **Lithology** — rock type, textures, grain size, and geological descriptions
- **Alteration** — dominant alteration type, intensity, and mineral assemblages
- **Mineralization** — logged sulphide percentages (Py, CCP, Bn)
- **Collar** — drillhole location, orientation, and total depth

### 4. Geological Background

The MPD project is located within the Nicola Belt, a terrane known for 
Cu-Au porphyry deposits. Porphyry systems typically exhibit:

- Broad halos of weak copper mineralization with trace Cu present 
  throughout background host rocks
- Localised higher-grade intervals associated with sulphide minerals 
  (chalcopyrite, bornite, pyrite)
- Strong lithology–alteration–mineralization relationships that define 
  predictable zonation patterns from the porphyry core outward
- Skarn development at contacts between intrusive rocks and reactive 
  host lithologies, often associated with elevated Cu-Au grades

These characteristics directly influence how the dataset is structured 
and interpreted throughout this project.

### 5. Key Findings

The 2023 MPD drill program comprised 26 drillholes across two target 
areas — AXE and MPD. Analysis of the program identified:

- **Monzonite (MM) with skarn alteration (SKN)** as the strongest 
  grade host, returning average Cu grades exceeding 0.90%
- **AXE-23-011** as the standout hole — 0.55% Cu p90, 73% skarn 
  lithology proportion, and the highest composite representativeness 
  score in the program
- A high-grade Cu-Au zone concentrated between **1200–1300m elevation** 
  at Drill Sites B and C, open at depth and along strike
- **MPD-23-005** showing an elevated Au signature relative to Cu — 
  potentially indicating a distinct mineralization style or distal 
  epithermal overprint worth further investigation

### 6. Validation Against Kodiak's Published Interpretation

Findings from this project were compared against Kodiak Copper's published geological interpretation in the 2023 ARIS assessment report (BC ARIS Report 42616) to assess how well the data-driven analysis aligns with expert geological interpretation.

**Where the analysis agrees with Kodiak:**
- AXE-23-011 identified as the standout hole — consistent with Kodiak's reported intercept of 0.49% Cu over 254m
- AXE-23-008 ranked second — consistent with Kodiak's reported 1.17% Cu over 39m shallow intercept
- West Zone (AXE target) outperforms Man Zone (MPD target) — 9 of 10 representative holes from AXE, consistent with Kodiak's conclusions
- MPD-23-005 Au-dominant signature noted — consistent with Kodiak's reported 0.25 g/t Au over 154m

**Where this analysis goes further:**
- Quantified SKN + MM as the strongest grade combination dataset-wide (>0.90% avg Cu across 68 intervals) — Kodiak describes this qualitatively hole by hole but does not prove it statistically
- Identified alteration intensity as a systematic grade predictor — not addressed in Kodiak's report
- Identified bornite as a dataset-wide Au indicator beyond MPD-23-005
- Defined 1200–1300m elevation as a specific target horizon — more actionable than Kodiak's general reference to depth potential
- Built a reproducible composite hole scoring framework applicable to future drill programs
